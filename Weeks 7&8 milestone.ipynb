```python
import json
import requests
import pandas as pd
```

```python
df=pd.read_json('https://covidtracking.com/api/states')
```

```python
df.head()
```

```python
df = df.fillna(" ")
df.isnull().sum()
```

```python
df.to_csv('covidAPI.csv',index=False)
```

```python
df = df.drop('hash', 1)
df = df.drop('commercialScore',1)
df = df.drop('negativeRegularScore',1)
df = df.drop('negativeScore',1)
df = df.drop('positiveScore',1)
df = df.drop('score',1)
df = df.drop('grade',1)
```

```python
df = df.drop([3,12,27,42,50], axis=0)
```

```python
indexNames = df[df['dataQualityGrade'] == 'D' ].index
```

```python
df.drop(indexNames , inplace=True)
```

```python
indexNames = df[df['dataQualityGrade'] == 'C' ].index
df.drop(indexNames , inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
```
