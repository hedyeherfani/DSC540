```python
def strip_whitespace(s):

    return s.strip()
```

9) Use the apply method to apply this function to all the columns with string values, create a new column, copy the values from this new column to the old column, and drop the new column.

```python
# Education column

df2['education_stripped']=df['education'].apply(strip_whitespace)

df2['education']=df2['education_stripped']

df2.drop(labels=['education_stripped'],axis=1,inplace=True)

# Occupation column

df2['occupation_stripped']=df['occupation'].apply(strip_whitespace)

df2['occupation']=df2['occupation_stripped']

df2.drop(labels=['occupation_stripped'],axis=1,inplace=True)
```
