```python
import sqlite3
```

```python
my_connection = sqlite3.connect('Covid')
c = my_connection.cursor()
query = """CREATE TABLE minis (fig_key TEXT PRIMARY KEY,
              'set' TEXT, num TEXT, name TEXT, age FLOAT, orig FLOAT,
              rarity TEXT, repaint INTEGER, size TEXT);"""
c.execute(query)
```

```python
droptablestatement="drop table minis"
c.execute(droptablestatement)
```
